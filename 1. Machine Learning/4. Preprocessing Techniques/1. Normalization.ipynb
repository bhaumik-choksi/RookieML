## Problem
Some features may have a larger range, compared to other features. During training, these features with a larger range can "dominate" the features with a smaller range, or features with a numerically smaller value. 

## Solution
We must make sure that all features have similar ranges. Normalization converts a given distribution of values to a normal distribution. A normal distribution is symmetrical around its center. For a normal distribution, mean = median = mode

## Example
Let us take two features, age (0-99) and salary (0-10,000). <br>
We will randomly generate 100 values of each feature, within the given range. <br>
We will then plot the data to see how our data is distributed. <br>

```python
import numpy as np
from matplotlib import pyplot as plt
import matplotlib as mpl

# We use the numpy library to generate random values in a given range
age = np.random.randint(0,99, 100, dtype=np.int32)

salary = np.random.randint(0,100000 , 100, dtype=np.int32)

plt.title("Before Normalization")
plt.plot(age, c="red")
plt.plot(salary, c="blue")
plt.show()
```

```python
# The np.mean() method returns the mean of a numpy array
# The np.std() method returns the standard deviation of a numpy array
age = ( age - np.mean(age) ) / np.std(age)
salary = ( salary - np.mean(salary) ) / np.std(salary)

plt.title("After Normalization")
plt.plot(age, c="red")
plt.plot(salary, c="blue")
plt.show()
```

As you can see, before normalization, <em style="color:red"> age </em> is insignificant compared to <em style="color:blue"> salary </em>. <br>
After normalization, both features have equal weight.
