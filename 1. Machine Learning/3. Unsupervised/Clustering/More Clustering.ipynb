We can see how simple distance based k-means clustering is ineffective at identifying clusters. 

We need a more robust technique that takes into consideration the connectivity and density of points. 

## Enter DBSCAN
Density-based Spatial Clustering of Applications with Noise

#### What is it?
DBSCAN relies on the density of point to perform clustering, instead of the distance from centroids. <br>
Since it does not depend on the centroids, you do not need to specify the number of clusters. <br>
However, you do need to specify two additional parameters: 
1. Epsilon: The maximum distance between two data points to be considered in the same neighborhood
2. min_samples : The minimum amount of data points in a neighborhood to be considered a cluster

Here's the algorithm: <br>
![algo](https://image.slidesharecdn.com/dbscanalgorithompresentation-161025075009/95/dbscan-algorithom-17-638.jpg?cb=1477382075)

```python
# Let's look at how to implement this using Scikit-learn

# We start by generating the clusters
import numpy as np
from matplotlib import pyplot as plt
from sklearn.datasets import make_moons


x,y = make_moons()
x = x.T # Convert (x,y) tuples to array_of_x and array_of_y coordinates
arr_x = x[0]
arr_y = x[1]
plt.scatter(arr_x, arr_y)
plt.show()

x = x.T # Convert back to list of (x,y) tuples
```

```python
from sklearn.cluster import KMeans, DBSCAN

dbscan = DBSCAN(eps=0.5, min_samples=10).fit(x)
kmeans = KMeans(n_clusters=2).fit(x)
```

### K-means output

```python
plt.scatter(arr_x, arr_y, c=kmeans.labels_)
plt.show()
```

### DBSCAN output

```python
plt.scatter(arr_x, arr_y, c=dbscan.labels_)
plt.show()
```

As the results show, DBSCAN was better at identifying the clusters, as compared to K-means clustering.
