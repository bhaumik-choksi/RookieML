# What is clustering?
Clustering is a type of unsupervised learning algorithm. Unlike supervised learning algorithms that predict a value or a class, unsupervised algorithms merely find similarity between various inputs and group them together based on this "similarity".
![image](https://home.deib.polimi.it/matteucc/Clustering/tutorial_html/images/clustering.gif)


## How is it different from classification? 
Classification categorizes an input into one of many pre-defined classes. These classes are also given as a part of the input. On the other hand, clustering merely finds out which inputs are similar to each other. For clustering, no information regarding pre-defined classes is given. 

In interactive demo is available [here](http://www.onmyphd.com/?p=k-means.clustering)

# Simple Example :)
For clustering, instead of defining our own data, we'll use the Iris datasets which is included in scikit-learn. The Iris dataset contains data about the length of various parts of the Iris flower, and the specie they belong to. Since we're performing clustering, and not classification, we'll ignore the specie for now.

```python
from sklearn import datasets
import sklearn.cluster as cl

Iris = datasets.load_iris()

# We take only 2 features for now so that we can visualize it on a 2D plot
X = Iris.data[:, :2]

# We don't need the target since this is unsupervised learning
# Y = Iris.target

from matplotlib import pyplot as plt
plt.scatter(X[:, 0], X[:, 1]) # Plotting 1 features on each axis
plt.show()
```

#### In order to perform Kmeans clustering, we need to specify how many clusters we want to make. 
Why? Can't the model just figure it out? 
#### Well... it could. But may happen that every datapoint is given a new cluster, so every point is a cluster on it's own
#### ...which seems quite pointless -_-
This happens because Kmeans tries to minimize the distance between a point and the centroid of the cluster it belongs to. So if every point were a cluster, the centroid of that cluster would be the point itself, and hence the distance would be zero, which is the ideal case. But such a clustering result would be quite useless.
#### There is a method called the [elbow method](https://goo.gl/4JojYT) which tries to accurately predict the number of clusters. You might want to check it out :)

```python
km = cl.KMeans(n_clusters=3) # Initializing a model which makes 3 clusters
km.fit(X) # Fitting data into the cluster
clusters = km.labels_ # Extracting the cluster labels after the clustering is complete

# Plotting the same data, but this time with color labels
# These color labels are nothing but cluster numbers predicted by the model
plt.scatter(X[:, 0],X[:, 1], c=clusters)
plt.show()
```

```python
print(clusters) # Raw output of the model is just cluster numbers
```
